```python
%%writefile locustfile.py
from locust import HttpUser, TaskSet, task, between

class UserBehavior(TaskSet):
    host = "http://172.28.224.1"    # O endereço do servidor Nginx a ser testado
    wait_time = between(5, 15)      # O tempo de espera entre as tarefas para cada usuário virtual é um valor aleatório entre 5 e 15 segundos
    
    @task
    def index(self):
        self.client.get("/")

    @task
    def blog_post_with_large_image(self):
        self.client.get("/2023/06/17/post1/")

    @task
    def blog_post_with_text(self):
        self.client.get("/2023/06/15/post2/")

    @task
    def blog_post_with_small_image(self):
        self.client.get("/2023/06/17/post3/")

class WebsiteUser(HttpUser):
    tasks = [UserBehavior]
    wait_time = between(5, 15)
```

# Versão 3 instâncias

```python
# Cria um DataFrame vazio para armazenar todos os dados
all_data = pd.DataFrame()

# Itera por todos os arquivos CSV
for file in csv_files:
    # Lê o arquivo CSV em um DataFrame
    df = pd.read_csv(file)

    # Adiciona colunas para o número de usuários e instâncias do WordPress
    df['users'] = int(file.split('_')[2])
    df['instances'] = int(file.split('_')[4][0])

    # Adiciona os dados ao DataFrame principal
    all_data = pd.concat([all_data, df])
    
# Agrupa os dados pelo número de usuários e instâncias do WordPress e calcula a mediana
grouped  = all_data.groupby(['users', 'instances'])["Median Response Time"].median().reset_index()
sizes    = all_data.groupby(['users', 'instances'])["Average Content Size"].median().reset_index()
requests = all_data.groupby(['users', 'instances'])["Requests/s"].median().reset_index()
failures = all_data.groupby(['users', 'instances'])["Failures/s"].median().reset_index()
```

# Versão Balanceador

```python
# Cria um DataFrame vazio para armazenar todos os dados
all_data = pd.DataFrame()

# Itera por todos os arquivos CSV
for file in csv_files:
    # Lê o arquivo CSV em um DataFrame
    df = pd.read_csv(file)

    # Adiciona colunas para o número de usuários e instâncias do WordPress
    df['users'] = int(file.split('_')[-2])

    # Adiciona os dados ao DataFrame principal
    all_data = pd.concat([all_data, df])
    
# Agrupa os dados pelo número de usuários e instâncias do WordPress e calcula a mediana
bal_users    = all_data.groupby(['users'])["Median Response Time"].median().reset_index()
bal_sizes    = all_data.groupby(['users'])["Average Content Size"].median().reset_index()
bal_requests = all_data.groupby(['users'])["Requests/s"].median().reset_index()
bal_failures = all_data.groupby(['users'])["Failures/s"].median().reset_index()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Criando uma lista única de 'users' e 'instances'
users = grouped['users'].unique()
# instances = grouped['instances'].unique()

# Largura das barras
bar_width = 0.25

# Criando um índice para cada grupo de 'users'
index = np.arange(len(users))

plt.figure(figsize=(12, 6))  # Ajuste o tamanho da figura

ax1 = plt.gca()  # Pega a referência do eixo atual (onde as barras são desenhadas)

# Criando um gráfico de barras para cada 'instance'
for i in instances:
    ax1.bar(index + bar_width*(i-1), 
            grouped[grouped['instances']==i]['Median Response Time'], 
            bar_width, 
            label=f'Instance {i}')

ax2 = ax1.twinx()  # Cria um novo eixo y com a mesma escala x

# Plotando 'Requests/s' e 'Failures/s' no eixo secundário
for i in instances:
    ax2.plot(index + bar_width*(i-1),
             requests[requests['instances']==i]['Requests/s'],
             marker='o',
             label=f'Requests/s Instance {i}')
    ax2.plot(index + bar_width*(i-1),
             failures[failures['instances']==i]['Failures/s'],
             marker='o',
             label=f'Failures/s Instance {i}')

# Adicionando rótulos, título, eixo x e y e uma legenda
ax1.set_xlabel('Users')
ax1.set_ylabel('Median Response Time')
ax2.set_ylabel('Requests/s and Failures/s')  # Adiciona um rótulo ao eixo y secundário
plt.title('Median Response Time by Users and Instances')
ax1.set_xticks(index + bar_width)
ax1.set_xticklabels(users)

# Combinando as legendas
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left', bbox_to_anchor=(1,1))

# Adicionando rótulos de dados nas barras
for i in range(len(users)):
    for j in range(len(instances)):
        y = grouped[(grouped['users'] == users[i]) & 
                    (grouped['instances'] == instances[j])]['Median Response Time'].values[0]
        ax1.text(i + bar_width*j, y, int(round(y, 2)), ha = 'center')

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```
